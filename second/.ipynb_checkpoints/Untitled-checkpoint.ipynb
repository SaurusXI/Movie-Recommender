# Content based filtering

```python
import dask.dataframe as dd
import dask.array as da
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from dask_ml.metrics.pairwise import linear_kernel
```

```python
links = dd.read_csv('../links.csv')
top_charts = pd.read_csv('../FIRST-top_charts.csv')
metadata = pd.read_csv('../movies_metadata.csv')
metadata.head()
```

```python
metadata['tagline'] = metadata['tagline'].fillna(' ')
metadata['overview'] = metadata['overview'].fillna(' ')
metadata['description'] = metadata['overview'].apply(lambda x: x.lower()) + ' ' + metadata['tagline'].apply(lambda x: x.lower())
metadata.head()
```

```python
vectorizer = TfidfVectorizer(analyzer = 'word', stop_words = 'english')
description_matrix = vectorizer.fit_transform(metadata['description'])
description_matrix.shape
```

```python
similarity_matrix = da.rechunk(linear_kernel(description_matrix, description_matrix), chunks = (100, 1))
indices = pd.Series(metadata.index, index = metadata['title'])
titles = pd.Series(metadata['title'], index = metadata.index)
```

```python
def get_recommendations(movie_name):
    index = indices[movie_name]
    scores = list(enumerate(similarity_matrix[index]))
    scores = sorted(scores, key = lambda x: x[1], reverse = True)[1:51]
    movie_indices = [i[0] for i in scores]
    recommendations = [titles[i] for i in movie_indices]
    return titles
```
